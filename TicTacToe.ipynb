```python
# Imports

import numpy as np
import random
```

```python
# Matrix is 3x3, filled with None data type

matx = np.full((3,3),None)
print(matx)
```

```python
# Logic for horizontal wins

matx[0,0] = 'x'
matx[0,1] = 'x'
matx[0,2] = 'x'

for i in range(len(matx)):
    for j in range(len(matx)-2):
        if matx[i,j] == 'x' and matx[i,j+1] == 'x' and matx[i,j+2] == 'x':
            print('Tic Tac Toe')

# Works
```

```python
# Logic for vertical wins

matx[0,0] = 'x'
matx[1,0] = 'x'
matx[2,0] = 'x'

for i in range(len(matx)-2):
    for j in range(len(matx)):
        if matx[i,j] == 'x' and matx[i+1,j] == 'x' and matx[i+2,j] == 'x':
            print('Tic Tac Toe')
        else:
            print('Nothing')
# Works
```

```python
# Logic for first diagonal wins

matx[0,0] = 'x'
matx[1,1] = 'x'
matx[2,2] = 'x'

for i in range(len(matx)-2):
    for j in range(len(matx)-2):
        if matx[i,j] == 'x' and matx[i+1,j+1] == 'x' and matx[i+2,j+2] == 'x':
            print('Tic Tac Toe')
        else:
            print('Nothing')
# Works
```
